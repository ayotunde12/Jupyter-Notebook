```python
from bs4 import BeautifulSoup

import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Nigeria'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table', class_ = 'wikitable sortable jquery-tablesorter')
```

```python
table1 = soup.find_all('table')[0]
```

```python
title1 = table1.find_all('th')
```

```python
Updatedtitle = [title.text.strip() for title in title1]
```

```python
Updatedtitle
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = Updatedtitle)
```

```python
df
```

```python
column_data1 = table1.find_all('tr')
```

```python
for row in column_data1[1:]:
    row_data = row.find_all('td')
    individual_data = [data.text.strip() for data in row_data]
    
    print(individual_data)
    
    length = len(df)
    df.loc[length] = individual_data
```

```python
df
```
